# Medidas de aceleração 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from math import sqrt
```

```python
custom_names = ['time', 'ax', 'ay', 'az', 'abs']
df = pd.read_csv("data/walk_george1.csv", header=0, names=custom_names) 
df.head()
```

```python
df.time
```

```python

fig, axs = plt.subplots(nrows=3, ncols=1, sharex=True, sharey=True)

axs[0].plot(df.time, df.ax, c='r')
axs[0].set_title('$a_x$', loc='right', x=0.9, y=0.7)
axs[1].plot(df.time, df.ay, c='b')
axs[1].set_title('$a_y$', loc='right', x=0.9, y=0.7)
axs[1].set_ylabel('$m/s^2$')
axs[2].plot(df.time, df.az, c='g')
axs[2].set_title('$a_z$', loc='right', x=0.9, y=0.7)
axs[2].set_xlabel('Tempo (s)');

fig.savefig('accel.png')

```
